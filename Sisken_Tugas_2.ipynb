```python
%matplotlib inline

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import control.matlab as control

G = control.tf([1],[10*1, 1*10/16, 1])
print(G)
```

```python
y,t = control.step(G)
plt.plot(t,y)
plt.xlabel('Time')
plt.title('Step Response')
```

```python
# impulse
y,t = control.impulse(G)
plt.plot(t,y)
plt.xlabel('Time')
plt.title('Step impulse')
```
